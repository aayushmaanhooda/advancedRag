HYBRID

```python
from langchain_community.retrievers import BM25Retriever
from langchain_community.vectorstores import FAISS
from langchain_classic.retrievers import EnsembleRetriever
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()
```

```python
# 1. BM25 retriever (keyword search)
bm25_retriever = BM25Retriever.from_documents(documents)
bm25_retriever.k = 3

# 2. Vector retriever (semantic search)
embeddings = OpenAIEmbeddings()
vectorstore = FAISS.from_documents(documents, embeddings)
vector_retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
# 3. Ensemble (hybrid) retriever
ensemble_retriever = EnsembleRetriever(
    retrievers=[bm25_retriever, vector_retriever],
    weights=[0.4, 0.6]  # Equal weight to both
)
```

```python
# 4. Query
results = ensemble_retriever.invoke("My goal here is to leanr python programming, so give me few docs on that")

for doc in results:
    print(doc.page_content)
```
